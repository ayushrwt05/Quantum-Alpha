

## Derivatives

A derivative is a financial contract whose value is derived from an underlying asset such as an index, stock, commodity, or currency. These instruments are widely used for hedging risk, speculation, and portfolio management.

For example, NIFTY options are derivatives because their prices depend on movements in the NIFTY index.

## Call Option

A call option gives the buyer the right, but not the obligation, to buy an asset at a predetermined price called the strike price.

Traders typically buy call options when they expect the market to rise. If the market price increases above the strike price, the call option can become profitable.

## Put Option

A put option gives the buyer the right, but not the obligation, to sell an asset at a predetermined strike price.

Put options are commonly used when traders expect the market to fall. During market crashes, put option values can increase significantly.

## ITM, ATM, and OTM Options

Options are categorized based on their relationship to the current market price.

- In-The-Money (ITM): The option has intrinsic value and would be profitable if exercised.
- At-The-Money (ATM): The strike price is close to the current market price.
- Out-of-The-Money (OTM): The option has no intrinsic value and requires favorable price movement to become profitable.

ATM options are generally the most liquid and actively traded.

## Volatility

Volatility refers to the degree of variation in asset prices over time. It represents uncertainty or risk related to price movements.

High volatility indicates large and rapid price fluctuations, while low volatility indicates stable market behavior. In quantitative finance, volatility is often measured using the standard deviation of returns.

Volatility is often higher during market crises due to panic selling, liquidity shortages, and uncertainty about future economic conditions.

## India VIX

India VIX is a volatility index that measures market expectations of volatility over the next 30 days. It is derived from NIFTY option prices.

Higher VIX levels indicate increased fear and uncertainty in the market, while lower levels indicate stable conditions.

Typical interpretation:

- VIX around 12–15 → calm market
- VIX around 20–25 → rising uncertainty
- VIX above 40 → crisis conditions

## COVID Crash Observation (2020)

The NIFTY crash during March 2020 occurred as a sequence of sharp negative shocks rather than a smooth decline.

Volatility increased significantly during this period, leading to large daily price movements and gap openings. Prior to the crash, the market exhibited relatively low volatility and sideways movement.

Large gap-down openings during the crash indicate sudden information shocks and risk re-pricing in financial markets.

The recovery phase showed continued high volatility along with strong upward movements, demonstrating that market recoveries after crises are often unstable and uneven.

## Regime Concept in Financial Markets

Financial markets move through different regimes rather than following a constant trend.

Common regimes include:

- Low volatility trending regime
- High volatility panic regime
- Recovery or reversal regime

Understanding these regimes is essential for designing adaptive trading strategies. Reinforcement learning models can potentially identify regime shifts and adjust trading decisions dynamically.
