 ## Market Crash Behaviour & Regime Intelligence Framework


## 1. Market Behaviour Observation

During the February–March 2020 period, the NIFTY index experienced a rapid transition from a relatively stable regime into a panic-driven crash regime. Price action was characterized by consecutive large bearish candlesticks, frequent gap-down openings, and sharp expansion in daily trading ranges.

Prior to the crash, the market exhibited lower realized volatility and largely sideways movement, indicating temporary regime stability. The onset of the COVID-related uncertainty triggered a sudden repricing of risk across financial markets, resulting in accelerated downside momentum and liquidity withdrawal.


## 2. Drawdown Behaviour Observation

During the COVID-19 crash phase, the NIFTY index experienced an exceptionally rapid drawdown, with peak-to-trough declines unfolding over a very short time horizon. The speed of capital erosion during this regime was significantly higher than the pace of subsequent recovery.

Such asymmetric market behaviour highlights that downside risk materializes abruptly, whereas recovery typically occurs in a more gradual and uneven manner. This creates a temporary but substantial impairment in portfolio value for passive investors maintaining full market exposure.

Additionally, elevated volatility during drawdown phases increases the probability of behavioural errors such as panic selling or mistimed re-entry, further impacting realized investment outcomes.

These observations suggest that strategies capable of dynamically adjusting exposure during accelerating drawdown regimes may achieve improved capital preservation and more stable long-term compounding relative to static allocation approaches.


## 3. Market Microstructure Insights

- Significant expansion in intraday price ranges during the crash phase  
- Persistent negative momentum across multiple trading sessions  
- Gap-down openings indicating information shocks and liquidity vacuum  
- Increased probability of cross-asset correlation spikes during stress regimes  
- Elevated realized volatility even during the early recovery phase  

These observations suggest that market dynamics during crisis periods differ materially from normal trading environments.


## 4. Regime Interpretation

The crash regime represents a high-risk environment for passive exposure strategies due to deep drawdowns and volatility clustering. In such conditions, static portfolio allocation may lead to substantial temporary capital erosion.

Conversely, adaptive or systematic positioning strategies that dynamically adjust exposure based on prevailing market conditions may achieve improved risk-adjusted outcomes.

Recovery regimes following crashes are often unstable and uneven, reflecting gradual restoration of investor confidence and liquidity.


## 5. Feature Hypothesis for Reinforcement Learning Agent

To enable regime-aware decision making, the following measurable indicators may be incorporated into the RL state representation:

- Rolling realized volatility (e.g., 20-day standard deviation of returns)  
- Percentage drawdown from recent peak levels  
- Short-term momentum indicators (e.g., 20–50 day returns)  
- Trend strength metrics (e.g., moving average slope or crossover signals)  
- Volatility acceleration or change-rate indicators  

These features may help the agent identify transitions between stable, trending, and stressed market environments.


## 6. Strategy Implications for Dynamic Portfolio Allocation

- Reduce long exposure during high-volatility panic regimes  
- Allow tactical short positioning when strong negative momentum persists  
- Gradually increase allocation as volatility contracts and trend stability improves  
- Limit aggressive trading during sideways or uncertain market phases to reduce whipsaw risk  

Such adaptive positioning may enhance long-term Sharpe ratio and reduce maximum drawdowns relative to passive index exposure.


## 7. Empirical Reflection

Empirical observations suggest that financial markets do not exhibit constant statistical properties. Instead, they evolve through regime-dependent dynamics characterized by shifts in volatility, liquidity, and trend persistence.

Therefore, reinforcement learning frameworks that incorporate regime-sensitive features may provide a viable approach for improving dynamic portfolio allocation decisions.

## 9. Quantitative Regime Threshold Framework

While qualitative interpretation of market regimes provides useful intuition, systematic trading models require quantifiable thresholds to classify market states.

Possible statistical definitions of regimes include:

- High Volatility Regime → Rolling realized volatility above the 90th percentile of historical observations  
- Crash Regime → Index drawdown exceeding 15% from recent peak levels  
- Bullish Trend Regime → Price persistently trading above medium-term moving averages (e.g., 50-day moving average) with positive momentum signals  
- Sideways Regime → Low realized volatility combined with near-zero momentum  

Such threshold-based definitions allow reinforcement learning agents to operate within structured state representations rather than subjective market interpretations.

This approach improves reproducibility and supports empirical evaluation of regime-aware trading behaviour.


## 10. Regime Transition Dynamics and Model Risk

Market regime shifts are often characterized by abrupt increases in volatility, breakdown of established trend structures, and temporary deterioration in signal reliability.

These transition periods represent elevated model risk for systematic strategies, as historical relationships between indicators and price behaviour may temporarily weaken.

For reinforcement learning systems, regime transitions introduce additional uncertainty in reward realization and policy stability. Agents must therefore learn adaptive behaviour not only within stable regimes but also during periods of structural market change.

Understanding transition dynamics is essential for designing robust trading environments and realistic reward functions.


## 11. Implications for Reinforcement Learning Policy Design

The reinforcement learning agent is expected to learn sequential decision policies that adapt to evolving market regimes.

Within stable trending environments, maintaining directional exposure may enhance return generation. Conversely, during high-volatility or crash regimes, reducing exposure or adopting defensive positioning may improve capital preservation.

Additionally, the agent must learn to navigate regime transition phases where signal clarity diminishes and market behaviour becomes less predictable.

Therefore, incorporating regime-sensitive state variables such as volatility measures, momentum indicators, and drawdown metrics is critical for enabling adaptive portfolio allocation decisions.
