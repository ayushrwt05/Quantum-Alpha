```python
import pandas as pd
import numpy as np

market = pd.read_csv("../data/market_master.csv")

market["Date"] = pd.to_datetime(market["Date"])
market = market.set_index("Date")
```

```python
state_features = [
    "mom_20_norm",        # direction
    "vol_signal_norm",    # risk
    "trend_signal_norm",  # structure
    "dd_signal_norm",     # stress
    "vix_signal_norm",    # volatility expectation
    "breadth_signal_norm" # participation
]
```

```python
state_data = market[state_features].copy()
state_data = state_data.dropna()

state_array = state_data.values.astype(np.float32)
state_array.shape

print(state_array.shape)
```

```python
print(state_data.columns)
```

```python
state_data.to_csv("../data/state_dataset.csv", index=True)
np.save("../data/state_array.npy", state_array)
```
