## Momentum in Financial Markets

Momentum refers to the empirical tendency of asset prices to continue moving in the same direction over short to medium time horizons.

This phenomenon has been documented across global equity markets, commodities, and currencies, making it one of the most robust systematic trading signals.

Momentum may arise due to behavioural biases, delayed information diffusion, institutional positioning, and trend-following market participants.


## Trend Persistence

Trend persistence describes the tendency of price trends to sustain themselves once established.

Strong trends are often characterized by:

- consistent higher highs and higher lows in bullish regimes  
- consistent lower highs and lower lows in bearish regimes  
- price trading persistently above or below key moving averages  

Persistent trends create opportunities for systematic strategies to capture directional returns.

## Momentum Failure Conditions

Momentum signals do not perform uniformly across all market environments.

Common momentum failure scenarios include:

- sudden volatility spikes causing trend reversals  
- sideways or range-bound markets producing false breakouts  
- macroeconomic shocks leading to regime shifts  
- crowded positioning leading to sharp unwind moves  

Understanding momentum failure regimes is critical for designing adaptive trading strategies.


## Bullish vs Bearish Momentum Behaviour

Momentum dynamics differ across market regimes.

Bullish momentum regimes often exhibit smoother trend progression, gradual corrections, and sustained capital inflows.

Bearish momentum regimes tend to be more volatile, characterized by sharp downside acceleration, frequent gap moves, and higher realized volatility.

These differences suggest that trading strategies may require asymmetric positioning logic depending on trend direction.


## Interaction Between Momentum and Volatility

Momentum effectiveness is closely linked to prevailing volatility conditions.

- Low to moderate volatility environments often support stable trend formation.
- Extremely high volatility environments may disrupt trend persistence and increase reversal risk.

Therefore, combining momentum signals with volatility indicators may improve regime-aware trading decisions.


## Potential Momentum Features for RL State Representation

Possible measurable indicators include:

- 5-day return
- 20-day return
- Moving average crossover signals
- Price distance from moving average
- Trend slope indicators
- Consecutive positive or negative return count

These features may help the reinforcement learning agent identify directional persistence and potential trend exhaustion.
