## Market Regime Classification Framework

Financial markets do not exhibit constant statistical behaviour. Instead, they transition between different regimes characterized by variations in volatility, trend persistence, liquidity conditions, and investor sentiment.

Classifying market regimes helps systematic strategies adapt exposure and improve risk-adjusted performance.


## Comparative Characteristics of Market Regimes

| Regime | Volatility | Momentum | Breadth | Liquidity | Strategy Bias |
|--------|-----------|----------|---------|-----------|---------------|
| Bull Trend | Moderate | Positive | Broad | Supportive | Long / Aggressive |
| Crash | Extremely High | Strong Negative | Broad Selling | Tightening | Defensive / Short |
| Sideways | Medium | Mixed | Narrow | Neutral | Low Exposure |
| Recovery | High but Falling | Positive but Unstable | Improving | Gradually Improving | Tactical Long |


## Behavioural Interpretation of Regimes

Crash regimes are typically driven by panic selling, liquidity withdrawal, and rapid repricing of risk.

Bull regimes emerge when macro stability, earnings growth, and capital inflows support sustained trend formation.

Sideways regimes often reflect uncertainty, valuation consolidation, or conflicting macro signals.

Recovery regimes occur when volatility begins to decline and breadth gradually improves following major drawdowns.


## Reinforcement Learning Strategy Implications

An adaptive reinforcement learning strategy may modify behaviour depending on detected market regimes.

- Increase exposure during stable bullish regimes.
- Reduce risk during volatility spikes.
- Allow tactical short positioning during panic-driven crashes.
- Avoid overtrading during sideways consolidation phases.

Such regime-aware behaviour may improve long-term Sharpe ratio and drawdown control.

## Crash Regime Illustration: COVID-19 Market Shock in NIFTY Index

```python
import yfinance as yf
import matplotlib.pyplot as plt

nifty = yf.download("^NSEI", start="2018-01-01", progress=False)

plt.figure(figsize=(12,5))
plt.plot(nifty.index, nifty["Close"])

plt.axvspan("2020-02-01", "2020-04-30", color="red", alpha=0.3)
plt.title("COVID Crash Regime Highlight")
plt.show()
```

#### Empirical Observation

The highlighted COVID crash period illustrates how rapid volatility expansion and sharp negative momentum can define a distinct stress regime in equity markets. Such visually identifiable phases support the hypothesis that regime transitions can be detected using quantitative indicators rather than discretionary judgment.


## Quantitative Regime Detection Framework

Market regimes can be approximated using combinations of measurable indicators such as rolling volatility, medium-term momentum, market breadth proxies, and macro-sensitive variables.

For example, extremely high realized volatility combined with strong negative momentum and deteriorating breadth may signal a crash regime. Conversely, moderate volatility with sustained positive momentum and improving participation may indicate a bullish trend regime.

By integrating multiple signal dimensions, reinforcement learning agents may infer latent market regimes and dynamically adapt exposure decisions without requiring explicit regime labels.
