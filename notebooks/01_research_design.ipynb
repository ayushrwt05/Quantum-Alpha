# Quantum Alpha

## Reinforcement Learning Strategy for Dynamic Portfolio Optimization During Market Volatility

## Research Background

Market volatility in emerging markets like India creates significant opportunities and risks for active trading strategies. Traditional buy-and-hold index investing often experiences large drawdowns during crisis periods such as the COVID-19 crash. Reinforcement Learning provides a framework where trading agents can dynamically adjust their decisions based on market conditions.

## Objective

The objective of this project is to design and evaluate a Deep Q-Network based trading agent that learns optimal long and short positioning in the NIFTY index using technical indicators, volatility signals, and macroeconomic variables.

## Hypotheses

1. Reinforcement learning trading strategies can achieve higher Sharpe ratios than passive NIFTY investing during volatility shocks.

2. The RL agent will learn to reduce exposure during high volatility regimes.

3. Combining technical indicators and macroeconomic signals improves trading performance.

## Evaluation Metrics

- Sharpe Ratio  
- Sortino Ratio  
- Maximum Drawdown  
- Calmar Ratio  

## Data Period and Methodology

Training Period: January 2019 – December 2022  
Testing Period: January 2023 – December 2025  

The reinforcement learning agent will be trained on historical market data during the training period and evaluated on unseen market conditions during the testing period to ensure robustness and avoid overfitting.

## Trading Universe

The trading agent will take dynamic long and short positions in the NIFTY index based on market signals. The strategy will not initially use derivatives such as options to maintain clarity and focus on directional alpha generation.
