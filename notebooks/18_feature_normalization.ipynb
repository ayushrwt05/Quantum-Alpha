```python
import pandas as pd
import numpy as np

market = pd.read_csv("../data/market_master.csv")

market["Date"] = pd.to_datetime(market["Date"])
market = market.set_index("Date")
```

```python
features = [
    "mom_5",
    "mom_20",
    "mom_diff",
    "vol_signal",
    "trend_signal",
    "trend_slope",
    "drawdown",
    "dd_signal",
    "vix_signal",
    "usd_mom",
    "crude_mom",
    "breadth_signal"
]
```

```python
market = market.dropna(subset=features)
```

```python
stats = {}

for col in features:
    mean = market[col].mean()
    std = market[col].std()
    stats[col] = {"mean": mean, "std": std}

import json
with open("../data/normalization_stats.json", "w") as f:
    json.dump(stats, f)
```

```python
for col in features:
    mean = stats[col]["mean"]
    std = stats[col]["std"]
    
    if std != 0:
        market[col + "_norm"] = (market[col] - mean) / std
    else:
        market[col + "_norm"] = 0
    
    market[col + "_norm"] = market[col + "_norm"].clip(-3, 3)
```

```python
market[[col + "_norm" for col in features]].describe()
```

```python
print(market[[col + "_norm" for col in features]].mean())
print(market[[col + "_norm" for col in features]].std())
```

```python
market.to_csv("../data/market_master.csv")
```
