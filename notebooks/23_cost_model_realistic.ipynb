```python
import pandas as pd
import numpy as np

market = pd.read_csv("../backtesting/market_backtest_v2.csv")

market["Date"] = pd.to_datetime(market["Date"])
market = market.set_index("Date")
```

```python
#Base Cost (Same)
base_cost = 0.0003   # 3 bps

# Volatility cost (no position_change here)
market["vol_cost"] = market["vol_signal"].abs() * 0.0005

# Total cost (apply position_change once)
market["total_cost"] = market["position_change"] * (
    base_cost + market["vol_cost"] + 0.0007
)

#Slippage Model
market["slippage"] = market["position_change"] * 0.0007

#Total Cost
market["total_cost"] = (
    market["position_change"] * base_cost
    + market["vol_cost"]
    + market["slippage"]
)

#Updated Strategy Returns
market["strategy_ret_v2"] = (
    market["position"].shift(1) * market["nifty_ret"]
    - market["total_cost"]
)

market = market.dropna(subset=["strategy_ret_v2"])
```

```python
market.to_csv("../backtesting/market_backtest_v3.csv")
```
