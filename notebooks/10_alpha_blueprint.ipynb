## Project Objective

The objective of the Quantum Alpha framework is to design a regime-aware reinforcement learning strategy capable of dynamically adjusting portfolio exposure in response to evolving market conditions.

Rather than predicting short-term price movements directly, the strategy aims to optimize risk-adjusted returns by adapting trading behaviour across different volatility and momentum regimes.

The primary benchmark for evaluation is passive exposure to the NIFTY index.


## Market Regime Hypothesis

Financial markets exhibit time-varying statistical behaviour characterized by shifts in volatility, trend persistence, liquidity, and cross-asset correlations.

These regime transitions influence the effectiveness of trading signals and risk exposure decisions.

The Quantum Alpha framework assumes that incorporating regime-sensitive signals into the reinforcement learning state representation may improve capital preservation during stress regimes and enhance return capture during stable trending environments.


## Feature Families for State Representation

The reinforcement learning agent will observe a combination of technical and macro-financial indicators designed to capture market regime dynamics.

#### Momentum Features
- Short-term returns (e.g., 5-day)
- Medium-term returns (e.g., 20-day)
- Moving average relationships
- Trend slope indicators

#### Volatility Features
- Rolling realized volatility
- Volatility acceleration
- Implied volatility proxy (India VIX)

#### Market Breadth Proxies
- Sector momentum divergence
- Participation strength signals

#### Macro Risk Indicators
- Oil price shocks
- Currency movements (USD/INR)
- Interest rate regime changes


## Reinforcement Learning State Vector Concept

The agent's decision-making will be based on a structured state vector containing normalized regime indicators.

Example conceptual state:

state_t = [momentum_signal, volatility_signal, breadth_signal, macro_signal]

This representation allows the agent to detect regime transitions and adapt exposure decisions dynamically.


## Action Space Design

The reinforcement learning agent will operate within a discrete action framework including:

- Increase long exposure
- Reduce exposure
- Maintain neutral positioning
- Initiate tactical short positioning

Such structured action choices enable regime-aware portfolio adjustment rather than static allocation.


## Reward Design Philosophy

The reinforcement learning agent will seek to maximize cumulative portfolio returns while incorporating penalties for excessive drawdowns and trading costs.

Conceptually:

Reward = Portfolio Return − Risk Penalty − Transaction Cost

This formulation encourages strategies that balance return generation with capital preservation.


## Expected Strategy Behaviour

The Quantum Alpha strategy is expected to:

- Increase allocation during stable bullish momentum regimes
- Reduce exposure during volatility spikes
- Allow tactical short positioning during panic-driven crashes
- Limit overtrading during sideways consolidation phases

Such adaptive behaviour may improve long-term Sharpe ratio and reduce maximum drawdowns relative to passive benchmark exposure.


## Research Evaluation Plan

Strategy performance will be evaluated using:

- Cumulative return comparison vs NIFTY
- Sharpe ratio
- Sortino ratio
- Maximum drawdown
- Regime-specific performance analysis

Backtesting will be conducted across multiple market environments including the COVID crash and subsequent recovery phases.


## Reinforcement Learning Trading Episode Structure

The trading environment will be structured as sequential time steps representing daily market observations.

At each time step, the agent observes the current regime-sensitive state vector, selects an exposure action, and receives a reward based on resulting portfolio performance.

Episodes may correspond to continuous historical periods, allowing the agent to learn exposure transitions across multiple market regimes including bullish trends, volatility shocks, and recovery phases.

This sequential interaction framework enables the reinforcement learning model to optimize dynamic allocation policies over time.
