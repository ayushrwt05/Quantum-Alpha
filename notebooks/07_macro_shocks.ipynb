## Macro Shock Transmission in Emerging Markets

Equity markets in emerging economies such as India are highly sensitive to macroeconomic variables including interest rates, commodity prices, and currency movements.

Shocks in these variables can alter liquidity conditions, risk appetite, and capital flows, thereby influencing market volatility and trend persistence.

Understanding macro transmission mechanisms is essential for designing adaptive systematic trading strategies.


## Liquidity Tightening and Market Volatility

When central banks tighten monetary policy, financial liquidity in the system declines. This often leads to increased volatility, weaker trend sustainability, and higher drawdown probability in equity markets.

Conversely, accommodative policy environments may support trend persistence and risk-taking behaviour.


## Potential Macro Features for RL State Representation

Possible macro indicators include:

- Crude oil price momentum
- USD/INR trend or volatility
- Interest rate change signals
- Global risk proxies such as VIX

These variables may help reinforcement learning agents anticipate regime shifts beyond purely technical market signals.




## Macro Shock Transmission Proxy: Crude Oil vs Equity Momentum

```python
import yfinance as yf
import matplotlib.pyplot as plt

nifty = yf.download("^NSEI", start="2018-01-01", progress=False)
oil = yf.download("CL=F", start="2018-01-01", progress=False)

nifty_ret = nifty["Close"].pct_change().rolling(20).mean()
oil_ret = oil["Close"].pct_change().rolling(20).mean()

plt.figure(figsize=(12,5))
plt.plot(nifty_ret, label="NIFTY Momentum")
plt.plot(oil_ret, label="Crude Oil Momentum")
plt.legend()
plt.title("Macro Shock Proxy: Oil vs Equity Momentum")
plt.show()
```

#### Empirical Observation

The comparison between rolling momentum in crude oil and the NIFTY index highlights periods where macro-driven commodity trends diverge from equity market direction. Such divergence may indicate emerging macro stress conditions that could weaken equity trend stability.

Aligned movements between macro variables and equity performance may reflect supportive liquidity environments and stronger risk sentiment.

## Strategy Implications of Macro Shock Transmission

Macroeconomic shocks can alter market regimes by influencing liquidity availability, investor risk appetite, and capital flow dynamics. Sudden adverse macro developments may lead to increased volatility, breakdown of existing trends, and heightened drawdown risk.

Conversely, supportive macro environments may reinforce trend persistence and improve signal reliability for systematic strategies.

Incorporating macro-sensitive indicators into the reinforcement learning state representation may enable the agent to anticipate regime transitions and adjust exposure proactively rather than reacting solely to price-based signals.

Macro indicators may exhibit lagged relationships with equity regimes, requiring careful feature normalization and timing alignment in RL implementation.
